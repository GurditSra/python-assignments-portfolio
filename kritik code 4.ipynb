```python
import numpy as np
import matplotlib.pyplot as plt

# Declare variables
x, y = sp.symbols('x y')

# Example 1: Differentiating a function
func1 = x**2 + sp.sin(y)
deriv1_x = sp.diff(func1, x)  # Partial derivative with respect to x
deriv1_y = sp.diff(func1, y)  # Partial derivative with respect to y

print("Partial derivative of func1 with respect to x:")
print(deriv1_x)
print("Partial derivative of func1 with respect to y:")
print(deriv1_y)

# Example 2: Derivatives of another function
func2 = sp.exp(x) * sp.sin(y) + y**3
deriv2_x = sp.diff(func2, x)  # Partial derivative with respect to x
deriv2_y = sp.diff(func2, y)  # Partial derivative with respect to y

print("\nPartial derivative of func2 with respect to x:")
print(deriv2_x)
print("Partial derivative of func2 with respect to y:")
print(deriv2_y)

# Example 3: Gradient calculation of a function
g = x**2 * y + y**2 * x
grad_g_x = sp.diff(g, x)  # Gradient component with respect to x
grad_g_y = sp.diff(g, y)  # Gradient component with respect to y

# Display the gradient as a vector
gradient = sp.Matrix([grad_g_x, grad_g_y])
print("\nGradient of g(x, y):")
print(gradient)

# Substituting a specific point (1, -1) into the gradient
point = {x: 1, y: -1}
grad_value_x = grad_g_x.subs(point)
grad_value_y = grad_g_y.subs(point)

print("\nGradient values at (1, -1):")
print(f"({grad_value_x}, {grad_value_y})")

# Example 4: Second partial derivatives of a logarithmic function
h = sp.log(x**2 + y**2)
second_deriv_x = sp.diff(sp.diff(h, x), x)
second_deriv_y = sp.diff(sp.diff(h, y), y)
mixed_deriv_xy = sp.diff(sp.diff(h, x), y)
mixed_deriv_yx = sp.diff(sp.diff(h, y), x)

# Displaying the second partial derivatives as a matrix
hessian = sp.Matrix([[second_deriv_x, mixed_deriv_xy], [mixed_deriv_yx, second_deriv_y]])
print("\nHessian matrix of h(x, y):")
print(hessian)

print("\nThe mixed second partial derivatives are equal because the function is symmetric in x and y. This indicates that changing the differentiation order does not affect the result.")

# Example 5: Contour plot of a cubic function
j = x**3 - 3 * x * y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title(r'Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# Example 6: Oscillating and decaying function plot
f = sp.exp(-sp.sqrt(x**2 + y**2)) * sp.sin(4 * x) * sp.cos(4 * y)
f_func = sp.lambdify((x, y), f, 'numpy')

Z = f_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='coolwarm')
plt.colorbar()
plt.title(r'Contour plot of $f(x, y) = e^{-\sqrt{x^2 + y^2}} \sin(4x) \cos(4y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# Example 7: Contour plot of a combination function
k = sp.sin(x) * sp.cos(y) + (x**2) / 2 - (y**2) / 2
k_func = sp.lambdify((x, y), k, 'numpy')

Z = k_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='plasma')
plt.colorbar()
plt.title(r'Contour plot of $k(x, y) = \sin(x) \cos(y) + \frac{x^2}{2} - \frac{y^2}{2}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
